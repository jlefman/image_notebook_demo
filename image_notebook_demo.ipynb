```python
%pylab inline 
import cv2, numpy as np, PIL
from IPython.display import clear_output, display
from ipywidgets import interact, IntSlider
```

```python
def get_image():
    return PIL.Image.open('./data/ACHN 637 Control CC3_3Caption.png')

def blur_img(img, av, bv):
    av = av**bv
    kernel = np.ones((av,av),np.float32)/av**2
    img = cv2.filter2D(img,-1,kernel)   
    return img

a_slider = IntSlider(min=0, max=100, step=1, value=1)
b_slider = IntSlider(min=1, max=10, step=1, value=1)
img = get_image().resize((512,512))

@interact(av=a_slider, bv=b_slider)
def do_filt(av, bv):
    img_blur = blur_img(np.asarray(img), av, bv)
    imshow(PIL.Image.fromarray(img_blur))
    show()
    clear_output(wait=True)
```
